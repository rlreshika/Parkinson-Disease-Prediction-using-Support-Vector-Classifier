```python
import numpy as np
import pandas as pd
```

### Reading and Understanding the data

#### This data has been taken from UCI open Dataset and has been read using the read_csv function

```python
url = r"https://archive.ics.uci.edu/ml/machine-learning-databases/parkinsons/parkinsons.data"
df = pd.read_csv(url,encoding='latin1')
```

```python
df.head()
```

```python
df.drop("name",axis=1,inplace=True)
```

```python
df.shape
```

The name column is just the subject name and recording number which is not relevent to our model so removing number

```python
df.info()
```

All the columns are of numerical dtype

```python
df.describe()
```

```python
df_new = df.copy() # Creating a copy for comparision purpose
```

### Data PreProcessing

```python
# Calculating the null values present
df.isna().sum(axis=0)
```

<b>Insights:</b>
There are no null values in the dataset

```python
# Checking for the duplicate values
df_new.drop_duplicates(subset=None,keep='first',inplace=True)
```

```python
print(f"Shape before dropping duplicates: {df.shape} ")
print(f"Shape after dropping duplicates: {df_new.shape} ")
```

There are no duplicates in the dataset as the shape before and after dropping the duplicate is same

```python
# Checking for the garbage values 
for col in df.columns:
    print(df[col].value_counts().sort_values(ascending=False))
```

There are no garbage values in the dataset

There are no categorical columns in the dataset, so no need to convert any columns into categorical columns

### Exploratory Data Analysis

```python
# Importing the visualization library
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
```

```python
df.columns
```

```python
# Splitting the data into X and y
X = df.drop("status",axis=1)
y = df.pop("status")
```

```python
sns.scatterplot(x = X.iloc[:,0],y = X.iloc[:,1],hue=y)
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(X.iloc[:,0], X.iloc[:,1], X.iloc[:,2],c=y,cmap="coolwarm",s=100, alpha=0.5)
ax.set_xlabel(X.iloc[:,0].name)
ax.set_ylabel(X.iloc[:,1].name)
ax.set_zlabel(X.iloc[:,2].name)
plt.tight_layout()
plt.show()
```

```python
# Normalizing the data before passing through LDA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit_transform(X)
```

```python
from sklearn.metrics import silhouette_score
```

```python
print(f"silhouette_score: {silhouette_score(X,y)}")
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2,random_state=0,n_iter=1000,verbose=1)
X_transformed_tsne = tsne.fit_transform(scaler)
```

```python
tsne_df = pd.DataFrame({'comp-1':X_transformed_tsne[:,0],'comp-2':X_transformed_tsne[:,1]})
tsne_df['label'] = y
tsne_df.head()
```

```python
sns.scatterplot(x='comp-1',y='comp-2',data=tsne_df,hue='label')
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
X_oversampled,y_oversampled = oversample.fit_resample(X,y)
```

```python
# Normalizing the data before passing through LDA
from sklearn.preprocessing import StandardScaler
scaler_oversampled = StandardScaler().fit_transform(X_oversampled)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2,random_state=0,n_iter=1000,verbose=1)
X_oversampled_tsne = tsne.fit_transform(scaler_oversampled)
```

```python
tsne_df = pd.DataFrame({'comp-1':X_oversampled_tsne[:,0],'comp-2':X_oversampled_tsne[:,1]})
tsne_df['label'] = y
sns.scatterplot(x='comp-1',y='comp-2',data=tsne_df,hue='label',cmap='Blues')
```

```python
print(f"The silhouette score after over sampling of data  : {silhouette_score(scaler_oversampled,y_oversampled)}")
```
